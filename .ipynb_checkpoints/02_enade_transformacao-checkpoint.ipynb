## Transformações

- CO_IES CODIGO INSTITUIÇÃO DE ENSINO
- CO_CATEGAD CODIGO CATEGORIA ADMINISTRATIVA
- CO_GRUPO CODIGO GRUPO DO CURSO

```python
# Criou um novo objeto(Coluna) para atribuir
enade['DESC_PUBLICA'] = ''

# Quais categorias são setor privado
enade.loc[
    enade.CO_CATEGAD.isin([118, 120, 121, 10005, 10006, 10007, 10008, 10009, 17634]),
    'DESC_PUBLICA'
] = 'Privado'
```

```python
# Quais categorias são setor público
enade.loc[
    enade.CO_CATEGAD.isin([93, 115, 116, 10001, 10002, 10003]),
    'DESC_PUBLICA'
] = 'Público'
```

```python
# Mostra quantidade de setor privado e público
enade.DESC_PUBLICA.value_counts()
```

```python
# Termo relativo
enade.DESC_PUBLICA.value_counts() / enade.shape[0]
```

```python
# Transformar a modalidade
enade.CO_MODALIDADE = enade.CO_MODALIDADE.replace({
    0: 'EAD',
    1: 'Presencial'
})
```

```python
enade.CO_MODALIDADE.value_counts()
```

```python
# Transforma região
enade.CO_REGIAO_CURSO = enade.CO_REGIAO_CURSO.replace({
    1: 'Norte',
    2: 'Nordeste',
    3: 'Sudeste',
    4: 'Sul',
    5: 'Centro-Oeste'
})
```

```python
# Quantidade de Quantidade por região
enade.CO_REGIAO_CURSO.value_counts()
```
